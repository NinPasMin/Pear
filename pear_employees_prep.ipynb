```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
pear = pd.read_csv('pear_employees.csv', skipinitialspace = True)
pear.head()
```

```python
pear.info(memory_usage = 'deep')
```

```python
pear['emp_no'] = pear['emp_no'].astype('Int32')
pear['salary'] = pear['salary'].astype('Int32')
```

```python
pear.isna().sum()
```

```python
pear['emp_no'].nunique()
```

```python
pear['to_date'].max()
```

```python
pear['to_date'] = pear['to_date'].replace('9999-01-01', '2099-01-01')
```

```python
pear['to_date'] = pd.to_datetime(pear['to_date'], format = '%Y-%m-%d', yearfirst = True)
```

```python
pear['title_to_date'] = pear['title_to_date'].replace('9999-01-01', '2099-01-01')
```

```python
pear['title_to_date'] = pd.to_datetime(pear['title_to_date'], format = '%Y-%m-%d', yearfirst = True)
```

```python
pear['salary_to_date'] = pear['salary_to_date'].replace('9999-01-01', '2099-01-01')
```

```python
pear['salary_to_date'] = pd.to_datetime(pear['salary_to_date'], format = '%Y-%m-%d', yearfirst = True)
```

```python
pear['birth_date'] = pd.to_datetime(pear['birth_date'], format = '%Y-%m-%d', yearfirst = True)
```

```python
pear['hire_date'] = pd.to_datetime(pear['hire_date'], format = '%Y-%m-%d', yearfirst = True)
```

```python
pear['salary_from_date'] = pd.to_datetime(pear['salary_from_date'], format = '%Y-%m-%d', yearfirst = True)
```

```python
pear['title_from_date'] = pd.to_datetime(pear['title_from_date'], format = '%Y-%m-%d', yearfirst = True)
```

```python
pear[pear['to_date']!='2099-01-01'].sort_values('to_date').tail()
```

```python
pear.info(memory_usage = 'deep')
```

```python
pear['age_while_hired_days'] = (pear['hire_date'] - pear['birth_date']).dt.days.astype('Int32')
pear.head()
```

```python
pear['age_while_hired_years'] = np.floor(pear['age_while_hired_days'] / 365.25).astype('Int16')
```

```python
pear.head(3)
```

```python
pear = pear.drop('age_while_hired_days', axis =1)
```

```python
pear.head(2)
```

```python
pear['dept_no'].unique()
```

```python
pear['dept_no'] = pear['dept_no'].str.replace('d', 'D')
```

```python
new_column_order = ['emp_no', 'name', 'birth_date', 'gender', 'hire_date', 'age_while_hired_years', 'dept_name', 'dept_no',
       'from_date', 'to_date', 'salary', 'salary_from_date',
       'salary_to_date', 'title', 'title_from_date', 'title_to_date']
```

```python
pear.columns = ['emp_no', 'name', 'birth_date', 'gender', 'hire_date',
       'age_while_hired_years', 'dept_name', 'dept_no', 'dept_from_date', 'dept_to_date',
       'salary', 'salary_from_date', 'salary_to_date', 'title',
       'title_from_date', 'title_to_date']
```

```python
pear.head(3)
```

```python
pear = pear.sort_values(by=['emp_no', 'salary_from_date'])
```

```python
pear['previous_salary'] = pear.groupby('emp_no')['salary'].shift(1)
pear.head(30)
```

```python
pear[['emp_no', 'name', 'salary', 'previous_salary']].iloc[14:21]
```

```python
pear['salary_increase_pct'] = (((pear['salary'] - pear['previous_salary'])/pear['previous_salary'])*100).round(2)
```

```python
pear[['emp_no', 'name', 'salary', 'previous_salary', 'salary_increase_pct']].iloc[14:21]
```

```python
pear = pear.loc[:,['emp_no', 'name', 'birth_date', 'gender', 'hire_date',
       'age_while_hired_years', 'dept_name', 'dept_no', 'dept_from_date',
       'dept_to_date', 'salary', 'salary_from_date', 'salary_to_date', 'previous_salary', 'salary_increase_pct', 'title',
       'title_from_date', 'title_to_date']]
```

```python
(pear
 .groupby(['title'])
 .agg({'salary': ['max','min','mean']})).round(2)
```

```python
pear['emp_status'] = np.where(pear['dept_to_date'] == '2099-01-01', 'Active', 'Left')
```

```python
pear.head()
```

```python
pear['current_date'] = pd.to_datetime('2002-09-30')
```

```python
pear.head(2)
```

```python
pear['years_experience'] = ((np.where(pear['emp_status'] == 'Active', 
                                      pd.to_timedelta(pear['current_date'] - pear['hire_date']).dt.days, 0))/365.5).round(1)
```
